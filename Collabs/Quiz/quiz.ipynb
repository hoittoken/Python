````python
def productFib(prod: int):
    def lazy_fib():
            f1, f2 = 0, 1
            while True:
                yield [f1*f2, [f1,f2]]
                f1, f2 = f2, f1 + f2

    m, nums = next(filter(lambda x: x[0] >= prod, (x for x in lazy_fib())))

    return [*nums, m == prod]
    

print(productFib(4895))
print(productFib(5895))
````

````python
import time
import numpy as np

times_for_raw = []
times_for_func = []

def fasterer():
    for i in range(100000):
        i ** 3

for k in range(1000):
    
    ts_start = time.time()

    for i in range(100000):
        i ** 3

    ts_raw = time.time() - ts_start
    times_for_raw.append(ts_raw)

    ts_func = time.time()

    fasterer()

    ts_for_func = time.time() - ts_func
    times_for_func.append(ts_for_func)
    
times_for_raw = np.array(times_for_raw)
times_for_func = np.array(times_for_func)

print(f'25 квантиль для глобальных вычислений: {np.quantile(times_for_raw, 0.25)}')
print(f'25 квантиль для вычислений под функциец: {np.quantile(times_for_func, 0.25)}')
print(f'Среднее для глобальных вычислений: {np.quantile(times_for_raw, 0.5)}')
print(f'Среднее для вычислений под функциец: {np.quantile(times_for_func, 0.5)}')
print(f'75 квантиль для глобальных вычислений: {np.quantile(times_for_raw, 0.75)}')
print(f'75 квантиль для вычислений под функциец: {np.quantile(times_for_func, 0.75)}')
````

````python
# нет, оно не работает в онлайн коллабе. Запуск из-под VS only :)
import tkinter as tk
from tkinter import *
from tkinter import messagebox
import random as rnd

questions = {
    "На пути водоём с пираньями - искупнёмся?": 1,
    "Впереди дорождые работы - даже в подземельях кладут плитку. Идём в обход?": 0,
    "Из соседней пещеры вкусно пахнет шашлычком - идём туда?": 0,
    "Получено письмо от Максима из SkillFactory. Прочесть?": 0,
    "Атакует стая диких (но пушистых) корги. Сражаться?": 1,
    "В 1 удаве 38 попугаев?": 0,
    "Мимо пролетела стая пингвинов. Идём за ними?": 1,
    "Опять дорожные работы: в этот раз кладут трубы, а плитку снимают. Идём в обход?": 0,
    "Погладить манула?": 0,
}

default_answers = ["да", "нет"]

answers = [[x for x in default_answers] for x in range(len(questions))]

question_label = None
answer_buttons = []
tk = None
frame = None
rws = 0
steps = 0

class question():
    def __init__(self, question_text: str, answers: list[str], correctIndex: int) -> None:
        self.question_text = question_text
        self.answers = answers
        self.correctIndex = correctIndex

    def is_valid(self, actual: str) -> bool:
        return self.answers[self.correctIndex] == actual


class cell():
    def __init__(self, question: question, coords: tuple) -> None:
        self.question = question
        self.coords = coords
        self.is_visited = False
        self.is_scouted = False

    def visit(self):
        global answer_buttons,steps

        if self.is_visited:
            return

        self.is_scouted = True

        question_label["text"] = self.question.question_text

        for ab in answer_buttons:
            ab.destroy()
        answer_buttons = []

        cb_strs = self.question.answers
        self.var = StringVar()
        self.var.set(cb_strs[0])

        for i, item in enumerate(cb_strs):
            b = Radiobutton(frame, text=item, variable=self.var, font=("Arial", 15), value=item, command=self.__check_answer)
            b.grid(row=rws+i+1, columnspan=3)
            answer_buttons.append(b)

        self.render()


    def render(self):
        r, c = self.coords[0], self.coords[1]
        bg = "black"
        cmd = lambda: self.visit()

        if self.is_scouted:
            bg = "gray"

        if self.is_visited:
            bg = "white"
            cmd = None

        b = Button(frame, text=' ', background=bg, border=1, height=5, width=15, command=cmd)
        b.grid(row=r, column=c)



    def __check_answer(self):
        global steps, answer_buttons
        actual_answer = self.var.get()
        if self.question.is_valid(actual_answer) is False:
            question_label["text"] = "Это провал."
            result = messagebox.showerror(title="Это провал", message="Попробуйте снова")
            if result is not None:
                for widgets in frame.winfo_children():
                    widgets.destroy()
                frame.destroy()
                run(tk)

        question_label["text"] = "идём дальше..."
        for ab in answer_buttons:
            ab.destroy()
        answer_buttons = []
        self.is_visited = True
        self.render()

        steps -= 1
        if steps <= 0:
            result = messagebox.showinfo(title="Ура!", message="Это победа!")
            if result is not None:
                tk.quit()
                return



class runner():
    def __init__(self, rows: int = 3, cols: int = 3) -> None:
        global rws, steps
        rws = rows
        self.rows = rows
        self.cols = cols
        self.grid = [[0 for c in range(cols)] for r in range(rows)]

        steps = self.rows*self.cols

        qKeys = list(questions.keys())
        rnd.shuffle(qKeys)
        reshuffle = len(qKeys) < rows*cols
        i = 0
        for r in range(self.rows):
            for c in range(self.cols):

                if reshuffle:
                    rnd.shuffle(qKeys)

                q = qKeys[i]
                ans = answers[list(questions.keys()).index(q)]
                correct = questions[q]

                self.grid[r][c] = question(q, ans, correct)

                if not reshuffle:
                    i += 1

    def run(self, tk_prev):
        global question_label, tk, frame
        tk = tk_prev
        if tk_prev is None:
            tk = Tk()
            tk.title("Умопомрачительное подземелье")
            tk.geometry("600x600")
        frame = Frame(tk)

        self.__render()

        question_label = Label(frame, text="Клик-клик.", wraplength=400, font=("Arial", 15))
        question_label.grid(row=self.rows, columnspan=self.cols*2)
        frame.pack(side="top", pady=10)

        tk.mainloop()


    def __render(self):
        for r in range(self.rows):
            for c in range(self.cols):
                cl = cell(self.grid[r][c], (r, c))
                cl.render()


def run(tk_prev):
    game = runner()
    game.run(tk_prev)


run(None)
````

````python
tokens = {

        0: 'ноль',           1: 'один',           2: 'два',
        3: 'три',            4: 'четыре',         5: 'пять',
        6: 'шесть',          7: 'семь',           8: 'восемь',
        9: 'девять',

        10: 'десять',        11: 'одиннадцать',   12: 'двенадцать',
        13: 'тринадцать',    14: 'четырнадцать',  15: 'пятнадцать',
        16: 'шестнадцать',   17: 'семнадцать',    18: 'восемнадцать',
        19: 'девятнадцать', 
        
        20: 'двадцать',      30: 'тридцать',      40: 'сорок',
        50: 'пятьдесят',     60: 'шестьдесят',    70: 'семьдесят',
        80: 'восемьдесят',   90: 'девяносто',

        100: 'сто',          200: 'двести',       300: 'триста',
        400: 'четыреста',    500: 'пятьсот',      600: 'шестьсот',
        700: 'семьсот',      800: 'восемьсот',    900: 'девятьсот'

}

adjust_thousand = {1: 'одна', 2: 'две'}

# поддерживаемые разрядности
forms = {
    1000:           ['тысяча', 'тысячи', 'тысяч'],
    1000000:        ['миллион', 'миллиона', 'миллионов'],
    1000000000:     ['миллиард', 'миллиарда', 'миллиардов'],
}

divisors_hundreds = [100, 10, 1]

divisors_form = sorted(list(forms.keys()), reverse=True)

divisors = divisors_form + divisors_hundreds


# делим и отнимаем
# каждому делителю соответствует максимум трехзначное число
def divide_num(num: int, divisors: list[int]) -> dict[int, int]:

    parts = {}
    for d in divisors:
        quot = num // d
        num -= quot*d
        parts[d] = quot

    return parts


# выделяем сотни, десятки и единицы из числа
def convert_triplet(num: int) -> list[int]:

    if num == 0:
        return []

    parts = divide_num(num, divisors_hundreds)
    result = []

    hundreds = parts[100]*100
    tens = parts[10]*10
    ones = parts[1]

    # если в 'десятках' действительно 10, то можно сложить с единицами и отбросить единицы
    # в массиве токенов уже лежат числа от 10 до 19
    if tens == 10:
        tens = tens + parts[1]
        ones = 0

    if hundreds > 0:
        result.append(hundreds)
    if tens > 0:
        result.append(tens)
    if ones > 0:
        result.append(ones)

    return result


# "тысяча, тысячи, тысяч"
# получаем термин в зависимости от делителя и последней цифры числа
def get_form(div: int, num: int) -> str:

    if div not in divisors_form:
        return None

    form = None

    f = forms.get(div, None)

    if f is None:
        return None

    if num == 1:
        form = f[0]
    if num in (2, 3, 4):
        form = f[1]
    if num == 0 or num > 4:
        form = f[2]

    return form


# валидация входных параметров
def check_arg(arg) -> int:

    result = None

    try:
        result = int(arg)
        if len(str(arg)) > len(str(divisors[0]))+2:
            print(f'"{arg}" is too long')
            result = None
    except:
        print(f'"{arg}" is not a number')

    return result


# трансфигуратор (трансфигурирует) :)
def transfigurator(input) -> str:

    num = check_arg(input)
    if num is None:
        return

    # исходное число, разбитое на группы по делителю (3 цифры в группе)
    parts = divide_num(num, divisors)

    buf = {}
    # опять делим, но уже каждую группу и получаем набор идентификаторов токенов
    for div, val in parts.items():
        buf[div] = convert_triplet(val)

    # если число просто ноль - сразу выводим
    if sum(list(parts.values())) == 0:
        return tokens[0]

    result = []

    tens = buf[10][0] if len(buf[10]) > 0 else 0
    ones = buf[1][0] if len(buf[1]) > 0 else 0

    skip_mul = False
    # опять складываем единицы с десятками если необходимо
    if tens == 1 and ones:
        skip_mul = True
        buf[10][0] = tens*10 + ones
        del buf[1]

    # формируем число прописью
    for div, nums in buf.items():

        is_divisor = div in divisors_form

        if len(nums) == 0:
            continue

        # если число соответствует разряду делителя (тысяча, миллион и т.д) или
        # если число в разряде десятков и до этого складывали десятки с единицами - берем токен по числу
        # иначе умножаем число на делитель и берем токен по произведению
        for num in nums:
            n = num if is_divisor or (skip_mul and div == 10) else num*div
            token = tokens[n] if div != 1000 or n not in (1, 2) else adjust_thousand[n]
            result.append(token)

        if is_divisor:
            form = get_form(div, nums[-1])
            if form is not None:
                result.append(form)

    return ' '.join(result)


# test
input = [
    0, 0.9, 1, '0077', '9877', 19877, '98_str_77',
    17, 19, 201, 425, 999,
    1000, 1100, 1110, 1111, 3751, 4802,
    12345, 54321, 76543, 90796, 999999,
    123321501, 785456789, 999999999, 19000000000,
    1234567890, 9999999999, 111987654321,
    999998760000000, 1900000000000,
    99999999999999999999999
]

for i in input:
    r = transfigurator(i)
    if r is not None:
        print(f'{r}   ({i})')
````

````python
frase = 'фраза, которую мы шифруем.'
print(f"будем шифровать: {frase}")

frase_list = list(frase)

def coding(data, key=0):
    """_самый надёжный шифратор_

    Args:
        data (_list_): _фраза (в виде списка) которую шифруем_
        key (int, optional): _ключ шифра_. Defaults to 0.
    """
    result = []
    for i in range(len(data)):
        result.append(ord(data[i]) + key)
    return(result)
    
coded = coding(frase, 15)
print(f"зашифровали: {encripted}")

def decoding(data, key=0):
    """_самый надёжный дешифратор_

    Args:
        data (_list_): _фраза (в виде списка) которую дешифруем_
        key (int, optional): _ключ шифра_. Defaults to 0.
    """
    result = []
    for i in range(len(data)):
        result.append(chr(round(data[i] - key)))
    return(''.join(result))

print(f"расшифровали без ключа: {decoding(coded)}")
print(f"расшифровали с ключём: {decoding(coded, 15)}")
````

````python
import random
# Ищем перебором чисел п порядку
result = []
# Проверять будем 1000 раз
for i in range(1000):
    # Загадываем случайное число [1, 100]
    target = random.randint(1, 100)
    # Будем считать попытки
    attempts = 0
    current_num = 0
    while True:
        # если не угадали - прибавляем единичку
        if current_num != target:
            current_num += 1
            attempts += 1
        # как только угадаем - рвём бесконечный цикл
        else:
            break
    # Запишем с какой попытки угадали число
    result.append(attempts) 
# Посчитаем среднее по числу попыткам
result_order = round(sum(result) / len(result))   
print(f"Простым перебором в среднем угадываем на {result_order} ходу")

# Простым перебором угадываем на 50 ходу
````

````python
def geron(num, acc=2):
    n = len(str(num))
    if num % 2 == 0:
        x_0 = 7 * 10 ** ((n-2)/2)
    else:
        x_0 = 2 * 10 ** ((n-1)/2)
        
    x_1 = (x_0 + num / x_0) / 2

    while True:
        x_2 = (x_1 + num / x_1) / 2
        if round(x_1, acc) == round(x_2, acc):
            break
        x_1 = x_2   
    return round(x_2)

print(geron(6))
print(geron(9))
print(geron(100000000))
````

````python
import numpy as np
from numpy import sin,pi
import matplotlib.pyplot as plt
m=[[3,4],[5,4],[5,6],[9,8]] # отношение круговых частот
plt.figure(1)
for i in m:         
         a=i[0]
         b=i[1]
         d=0.5*pi
         x=[sin(a*t+d) for t in np.arange(0.,2*pi,0.01)]
         y=[sin(b*t) for t in np.arange(0.,2*pi,0.01)]        
         if m.index(i)==0:
                  plt.subplot(221)
                  plt.plot(x, y, 'k') # график для различных отношений  a/b круговых частот                                
                  plt.grid(True)
         elif m.index(i)==1:
                  plt.subplot(222)
                  plt.plot(x, y, 'g')               
                  plt.grid(True)
         elif m.index(i)==2:
                  plt.subplot(223)
                  plt.plot(x, y, 'b') 
                  plt.grid(True)
         else:
                  plt.subplot(224)
                  plt.plot(x, y, 'r')                  
                  plt.grid(True )    
plt.show()
````
